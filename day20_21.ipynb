```python
>=3 = blockbuster
2->3 = superhit
1->2 = hit
0->1 = average
<=0 = flop

(movie['earning'] - movie['cost']) / movie['duration']
```

```python
movie['profit_ratio'] = (movie['earning'] - movie['cost']) / movie['duration']
movie.head()
```

```python
# for ind, row in movie.iterrows():
#     if row['profit_ratio'] >= 3 
#         if row['status'] == 'blockbuster':
#             print('True')
#         else:
#             print('False')
#     elif row['profit_ratio'] >= 2 
#         if row['status'] == 'superhit':
#             print('True')
#         else:
#             print('False')

def logic(row):
    profit_ratio = (row['earning'] - row['cost'])/row['duration']
    if profit_ratio >= 3:
        return row['status'] == 'blockbuster'            
    elif profit_ratio >= 2:
        return row['status'] == 'superhit'
    elif profit_ratio >= 1:
        return row['status'] == 'hit'
    elif profit_ratio >= 0:
        return row['status'] == 'average'
    else:
        return row['status'] == 'flop'

movie['validation'] = movie.apply(logic, axis=1)
movie.head()
```

```python
# Data Joining
df1 = pd.DataFrame({
    'id':[1,2,3,4,5],
    'name':['ajay','vijay','suresh','ramesh','vinesh']
})
df2 = pd.DataFrame({
    'name':['vijay','ramesh','ronny','vinesh','kanishk'],
    'salary':[10,12,24,9,11]
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2], axis=1)
```

```python
pd.merge(df1,df2, on='name')
```

```python
pd.merge(df1,df2, on='name',how='outer')
```

```python
r = pd.merge(df1,df2, on='name',how='left')
r
```
